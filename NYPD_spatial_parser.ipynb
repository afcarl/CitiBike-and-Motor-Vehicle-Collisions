```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt, datetime, statsmodels.formula.api as smf, matplotlib.dates as mdates
%matplotlib inline
plt.style.use('ggplot')
```

```python
def convertDate(d):
    dt = datetime.datetime.strptime(d, '%m/%d/%Y')
    return dt

df = pd.read_csv('data/NYPD_Motor_Vehicle_Collisions.csv', converters={'DATE': convertDate})
```

```python
c_inj_agg = df.groupby('DATE').sum()['NUMBER OF CYCLIST INJURED']
```

```python
df_ = df[(df['NUMBER OF CYCLIST INJURED']>0)|(df['NUMBER OF CYCLIST KILLED']>0)]
print len(df),len(df_)
```

```python
df_.head()
```

```python
df_.to_csv('data/Bicycle_Collisions.csv')
```

```python
df__ = df[df['NUMBER OF CYCLIST KILLED']>0]
df__.to_csv('data_/Bicycle_Deaths.csv')
```

```python
d = pd.Timestamp(datetime.datetime(2013, 7, 1))
```

```python
d1 = df_[df_['DATE']<d]
d2 = df_[df_['DATE']>=d]
```

```python
print len(d1),len(d2)
d1
```

```python
d1.to_csv('data/b4ct_collisions.csv')
d2.to_csv('data/afct_collisions.csv')
```

```python
inj_b4 = d1.groupby('ZIP CODE').sum()['NUMBER OF CYCLIST INJURED']
inj_af = d2.groupby('ZIP CODE').sum()['NUMBER OF CYCLIST INJURED']
```

```python
inj_b4.values
```

```python
df_b4 = pd.DataFrame(columns=['ZipCode','NumberofIncidents'])
df_b4['ZipCode'] = inj_b4.index
df_b4['NumberofIncidents'] = inj_b4.values
df_b4
```

```python
df_b4 = df_b4.sort(['NumberofIncidents'],ascending=False)
df_b4[:20]
```

```python
df_b4.to_csv('data_/b4_collision_zip.csv')
```

```python
df_af = pd.DataFrame(columns=['ZipCode','NumberofIncidents'])
df_af['ZipCode'] = inj_af.index
df_af['NumberofIncidents'] = inj_af.values
df_af = df_af.sort(['NumberofIncidents'],ascending=False)
df_af[:20]
```

```python
df_af.to_csv('data_/af_collision_zip.csv')
```
