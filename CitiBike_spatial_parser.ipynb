```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt, urllib2
from zipfile import ZipFile
from StringIO import StringIO
%matplotlib inline
plt.style.use('ggplot')
```

```python
bike_months = [
    '201501',
    '201504',
    '201507',
    '201510'
]
```

```python
frames = []
for m in bike_months:
    url = 'https://s3.amazonaws.com/tripdata/%s-citibike-tripdata.zip'%m
    r = urllib2.urlopen(url).read()
    myfile = ZipFile(StringIO(r))
    df_csv = myfile.open('%s-citibike-tripdata.csv'%m)
    frame = pd.read_csv(df_csv)
    frames.append(frame)

df = pd.concat(frames, ignore_index=True)
df
```

```python
start = df.groupby('start station id').count()['start station name']
end = df.groupby('end station id').count()['end station name']
```

```python
df_s = pd.DataFrame(columns=['StationID','NumberofTrips'])
df_s['StationID'] = start.index
df_s['NumberofTrips'] = start.values
df_s = df_s.sort(['NumberofTrips'],ascending=False)
df_s
```

```python
df_s['StationID'][:10].tolist()
```

```python
stations = pd.DataFrame(columns=['id','lat','lon'])
count = 0
for i in df_s['StationID'][:20].tolist():
    stations.loc[count,'id'] = i
    stations.loc[count,'lat'] = df[df['start station id']==i]['start station latitude'].values[0]
    stations.loc[count,'lon'] = df[df['start station id']==i]['start station longitude'].values[0]
    count += 1
stations
```

```python
df_sf = df_s.merge(stations,how='inner',left_on='StationID',right_on='id')
df_sf
```

```python
df_e = pd.DataFrame(columns=['StationID','NumberofTrips'])
df_e['StationID'] = end.index
df_e['NumberofTrips'] = end.values
df_e = df_e.sort(['NumberofTrips'],ascending=False)
stationss = pd.DataFrame(columns=['id','lat','lon'])
count = 0
for i in df_e['StationID'][:20].tolist():
    stationss.loc[count,'id'] = i
    stationss.loc[count,'lat'] = df[df['end station id']==i]['end station latitude'].values[0]
    stationss.loc[count,'lon'] = df[df['end station id']==i]['end station longitude'].values[0]
    count += 1
df_ef = df_e.merge(stationss,how='inner',left_on='StationID',right_on='id')
df_ef
```

```python
df_test = df_sf.merge(df_ef,how='outer',on='id')
df_test
```

```python
df_test.loc[20,'lat_x'] = df_test.loc[20,'lat_y']
df_test.loc[20,'lon_x'] = df_test.loc[20,'lon_y']
df_test
```

```python
df_test = df_test.rename(columns={'lat_x':'lat','lon_x':'lon'})
df_f = df_test.drop(['StationID_x','NumberofTrips_x','StationID_y','NumberofTrips_y','lat_y','lon_y'],axis=1)
df_f
```

```python
for i in df_f.index:
    if i==15:
        df_f.loc[i,'popular as'] = 'start'
    elif i==20:
        df_f.loc[i,'popular as'] = 'end'
    else:
        df_f.loc[i,'popular as'] = 'both start & end'
df_f
```

```python
df_f.to_csv('data_/cb_pop_locs.csv')
```
